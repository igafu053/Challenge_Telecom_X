#üìå Extracci√≥n

```python
import requests
import pandas as pd
```

‚úÖ Cargar los datos directamente desde la API utilizando Python.

```python
url = "https://raw.githubusercontent.com/igafu053/Challenge_Telecom_X/refs/heads/main/TelecomX_Data.json"
response = requests.get(url)
data = response.json()

telecomx_data = pd.DataFrame(data)
```

```python
print(telecomx_data.head())
```

```python
telecomx_data.info()
```

```python
#Verificar las columnas del archivo
columnas = list(telecomx_data.columns)
columnas
```

```python
#Extraer los datos desde los diccionarios anidados en nuevas columnas
customer = pd.json_normalize(telecomx_data['customer'])
phone = pd.json_normalize(telecomx_data['phone'])
internet = pd.json_normalize(telecomx_data['internet'])
account = pd.json_normalize(telecomx_data['account'])

#Concadenar el nuevo dataframe con el dataframe original (excluyendo las columnas anidadas originales)
datos_telecom = pd.concat([telecomx_data[['customerID', 'Churn']], customer, phone, internet, account], axis=1)

#Mostrar las primeras filas del nuevo dataframe
datos_telecom.head()
```

```python
#Verificar las columnas del archivo normalizado
columnas = list(datos_telecom.columns)
columnas
```

#### Diccionario de datos

- `customerID`: n√∫mero de identificaci√≥n √∫nico de cada cliente
- `Churn`: si el cliente dej√≥ o no la empresa
- `gender`: g√©nero (masculino y femenino)
- `SeniorCitizen`: informaci√≥n sobre si un cliente tiene o no una edad igual o mayor a 65 a√±os
- `Partner`: si el cliente tiene o no una pareja
- `Dependents`: si el cliente tiene o no dependientes
- `tenure`: meses de contrato del cliente
- `PhoneService`: suscripci√≥n al servicio telef√≥nico
- `MultipleLines`: suscripci√≥n a m√°s de una l√≠nea telef√≥nica
- `InternetService`: suscripci√≥n a un proveedor de internet
- `OnlineSecurity`: suscripci√≥n adicional de seguridad en l√≠nea
- `OnlineBackup`: suscripci√≥n adicional de respaldo en l√≠nea
- `DeviceProtection`: suscripci√≥n adicional de protecci√≥n del dispositivo
- `TechSupport`: suscripci√≥n adicional de soporte t√©cnico, menor tiempo de espera
- `StreamingTV`: suscripci√≥n de televisi√≥n por cable
- `StreamingMovies`: suscripci√≥n de streaming de pel√≠culas
- `Contract`: tipo de contrato
- `PaperlessBilling`: si el cliente prefiere recibir la factura en l√≠nea
- `PaymentMethod`: forma de pago
- `Charges.Monthly`: total de todos los servicios del cliente por mes
- `Charges.Total`: total gastado por el cliente

```python
import numpy as np
```

```python
#Reemplazar espacios en blanco en valores nan y convertir strings en numeros
datos_telecom['Charges.Total'] = datos_telecom['Charges.Total'].replace(' ', np.nan)
datos_telecom['Charges.Total'] = pd.to_numeric(datos_telecom['Charges.Total'], errors='coerce')
datos_telecom.tail()
```

```python
#Convertir el tipo de datos de la columna 'Charges.Total' en datos float64
datos_telecom['Charges.Total'] = datos_telecom['Charges.Total'].astype(np.float64)
datos_telecom.info()
```

```python
#Columnas pocos relevantes para el analisis de los datos
columnas_excluidas = ['gender','SeniorCitizen','Partner','Dependents','PaperlessBilling','PaymentMethod']
```

‚úÖ Identificar las columnas m√°s relevantes para el an√°lisis de evasi√≥n.

```python
#Filtrar los datos mas relevantes del archivo para el analisis
datos_telecom_filtrado = datos_telecom.drop(columns=columnas_excluidas)
datos_telecom_filtrado
```

‚úÖVerificar si hay problemas en los datos

```python
# Verificar valores ausentes
print("Valores ausentes por columna:")
print(datos_telecom_filtrado.isnull().sum())
```

```python
# Verificar filas duplicadas
print("\nN√∫mero de filas duplicadas:", datos_telecom_filtrado.duplicated().sum())
```

```python
# Verificar valores √∫nicos en columnas categ√≥ricas para inconsistencias
print("\nValores √∫nicos en columnas categ√≥ricas:")
for col in datos_telecom_filtrado.select_dtypes(include='object').columns:
    print(f"- {col}: {pd.unique(datos_telecom_filtrado[col])}")
```

```python
# Verificar la distribuci√≥n de las categor√≠as en columnas categ√≥ricas (opcional)
print("\nConteo de valores en columnas categ√≥ricas:")
for col in datos_telecom_filtrado.select_dtypes(include='object').columns:
    print(f"\n- {col}:")
    print(datos_telecom_filtrado[col].value_counts())
```

‚úÖEstandarizaci√≥n y transformaci√≥n de datos

```python
#Transformando nombres de columnas en formato legible

datos_telecom_filtrado.columns = datos_telecom_filtrado.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True).str.lower()
```
