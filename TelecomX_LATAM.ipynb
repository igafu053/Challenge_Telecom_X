#üìå Extracci√≥n

```python
import pandas as pd
```

‚úÖ Cargar los datos directamente desde la API utilizando Python.

```python
telecomx_data = pd.read_json('/content/drive/MyDrive/Aprendiendo_a_hacer_ETL_G8-ONE/Challenge_Telecom_X-AnaÃÅlisis_de_evasioÃÅn_de_cliente/Challenge_Telecom_X/TelecomX_Data.json')
telecomx_data.head(5)
```

```python
#Verificar las columnas del archivo
columnas = list(telecomx_data.columns)
columnas
```

```python
#Extraer los datos desde los diccionarios anidados en nuevas columnas
customer = pd.json_normalize(telecomx_data['customer'])
phone = pd.json_normalize(telecomx_data['phone'])
internet = pd.json_normalize(telecomx_data['internet'])
account = pd.json_normalize(telecomx_data['account'])

#Concadenar el nuevo dataframe con el dataframe original (excluyendo las columnas anidadas originales)
datos_telecom = pd.concat([telecomx_data[['customerID', 'Churn']], customer, phone, internet, account], axis=1)

#Mostrar las primeras filas del nuevo dataframe
datos_telecom.head()
```

```python
#Verificar las columnas del archivo normalizado
columnas = list(datos_telecom.columns)
columnas
```

‚úÖ Convertir los datos a un DataFrame de Pandas para facilitar su manipulaci√≥n.

```python
datos_telecom = pd.DataFrame(datos_telecom)
datos_telecom.head(5)
```

```python
#Convertir el ID del cliente en el indice del archivo
datos_telecom.set_index('customerID', inplace=True)
datos_telecom
```

```python
#Reemplazar espacios en blanco en valores nan y convertir strings en numeros
datos_telecom['Charges.Total'] = datos_telecom['Charges.Total'].replace(' ', np.nan)
datos_telecom['Charges.Total'] = pd.to_numeric(datos_telecom['Charges.Total'], errors='coerce')
datos_telecom.tail()
```

```python
#Convertir el tipo de datos de la columna 'Charges.Total' en datos float64
datos_telecom['Charges.Total'] = datos_telecom['Charges.Total'].astype(np.float64)
datos_telecom.info()
```
